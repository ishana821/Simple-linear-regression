Step 1: To import required packages and libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

Step 2: Read training and testing data from files

```python
df_train=pd.read_csv("C:\\Users\\hp\\Downloads\\AI ML TRAINING\\train.csv")
df_test=pd.read_csv("C:\\Users\\hp\\Downloads\\AI ML TRAINING\\test.csv")
#print (df_train, df_test)
```

Step 3: Load training and testing data in respective variables

```python
x_train=df_train['x']
y_train=df_train['y']
x_test=df_test['x']
y_test=df_test['y']
#print(x_train)
```

Step 4: Plot the relation between independent and dependent variable

```python
plt.figure(figsize=(10,10))
plt.scatter(x_train,y_train)
```

Step 5: Initialize the values of variables like learning rate (alpha), number of training instances (n), slope and intercept etc.

```python
n=700
alpha=0.0001
a_0=0.0
a_1=0.0
epochs=0
```

Step 6: Improvise the values of slope and intercept with the help of gradients.

```python
while (epochs<4000) : 
    y=a_0+a_1*x_train
    error=y-y_train
    a_0=a_0-a_0 - alpha*2*np.sum(error)/n
    a_1=a_1 - alpha*2*np.sum(error*x_train)/n
    epochs=epochs+1
#print (error)
```

Step 7: Predict the values for test data and check the goodness-of-fit measure with R2 Score.

```python
y_prediction=a_0+a_1*x_test
print('R2_score',r2_score(y_test,y_prediction))
```
